```python
df = df[df.Country == 'United Kingdom']
```

```python
null_values_count = df.isnull().sum()
print("Number of null values in each column:")
print(null_values_count)
```

```python
df = df.dropna(subset=['Description'])
```

```python
null_values_count = df.isnull().sum()
print("Number of null values in each column:")
print(null_values_count)
```

```python
qtt = df['Quantity'].unique()
print(qtt)
```

```python
# some spaces are there in description; need to remove else later operation it will create problem..
df['Description'] = df['Description'].str.strip()
```

```python
#some of transaction quantity is negative which can not be possible remove them.
df = df[df.Quantity >0]
```

```python
basket = (df[df['Country'] =="United Kingdom"] 
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo')) 
# Defining the hot encoding function to make the data suitable  
# for the concerned libraries 
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
  

  
```

```python
basket.shape
basket
```

```python
basket_encoded = basket.map(hot_encode) 
basket = basket_encoded 
  
```

```python
basket.head()
```

```python
# means 3 times in total number of transaction the item should be present.
frequent_itemsets = apriori(basket, min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_mlxtend
```
